```python
import pandas as pd

import seaborn as sns; sns.set()
import matplotlib
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
```

```python
path_to_data = 'https://gist.githubusercontent.com/Chepubelja/5b8596382611782513cab7e1c6c0257b/raw/27b9938e959286fa4d16d0d4efacd09e6fd1291c/rate.csv'
data = pd.read_csv(path_to_data)

data.fillna(0, inplace=True)
data['rate'] = pd.to_numeric(data['rate'], errors='coerce')
data = data.pivot("region", "year", "rate")

region_order = ["Zakarpattia", "Rivne", "Kyiv city", "Ivano-Frankivsk", "Volyn", "Chernivtsi", "Lviv", "Ternopil",
                "Sevastopol", "Crimea", "Odesa", "Kherson", "Mykolaiv", "Khmelnytskyi", "Zhytomyr", 
                "Zaporizhia", "Kyiv", "Vinnytsia", "Dnipropetrovsk", "Kharkiv", "Kirovohrad", "Cherkasy",
                "Donetsk", "Luhansk", "Poltava", "Sumy", "Chernihiv"]

data.index = pd.CategoricalIndex(data.index, categories=region_order)
data.sortlevel(level=0, inplace=True)

data
```

```python
# Creating custom colormap
import numpy as np

colors = ["#67001f", "#f7f7f7", "#053061"]
colors_rgb = [tuple(int(col.lstrip('#')[i:i+2], 16) / 255 for i in (0, 2, 4)) for col in colors]
binned_cm = LinearSegmentedColormap.from_list('mine', colors_rgb, N=35, gamma=1.7)
```

```python
# custom_cmap = mpl.colors.ListedColormap(['#67001f', '#f7f7f7', '#053061'])
binned_cm = LinearSegmentedColormap.from_list('mine', colors_rgb, N=100, gamma=1.64)

def get_alpha_blend_cmap(cmap, alpha):
    cls = plt.get_cmap(cmap)(np.linspace(0,1,256))
    cls = (1-alpha) + alpha*cls
    return ListedColormap(cls)

binned_cm = get_alpha_blend_cmap(binned_cm, 0.75)
```

```python
# Custom fonts preparation
mono_r = fm.FontProperties(fname='ubuntu-mono/UbuntuMono-R.ttf')
mono_b = fm.FontProperties(fname='ubuntu-mono/UbuntuMono-B.ttf')

plt.rcParams['ytick.labelright'] = plt.rcParams['xtick.labeltop'] = True
plt.rcParams['ytick.labelleft'] = plt.rcParams['xtick.labelbottom'] = False

# Set color of background
sns.set(rc={'axes.facecolor':'#F3F7F7', 'figure.facecolor':'#F3F7F7'})

grid_kws = {"height_ratios": (.015, .9), "hspace": .17}
fig, (cbar_ax, ax) = plt.subplots(2, figsize=(14, 9), gridspec_kw=grid_kws)
heatmapkws = dict(cmap = binned_cm, linewidths=0.0, vmin=-15.5, vmax=8.5)
ax = sns.heatmap(data, ax=ax, cbar_kws={"orientation": "horizontal"}, cbar_ax=cbar_ax,
                **heatmapkws);

# Put ticks inside (closer to axis)
cbar_ax.tick_params(direction='in')
cbar_ax.tick_params(left=False, bottom=False, top=False, right=False)

# Changed ticks to the only needed to display
locs = [-15.0, -8.0, 0., 8.0]
cbar_ax.xaxis.set_ticks(locs)

# Setting proper font size for cbar x tick labels
for tick in cbar_ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(12)
    
# Changing font family for cbar x tick labels
for label in cbar_ax.get_xticklabels():
    label.set_fontproperties(mono_r)

# Changing transparency of tick labels
plt.setp(cbar_ax.get_xticklabels(), color='#5D646F')

# Added title
ax.text(x=0.368, y=1.17, s='Population growth rate by region', color='#3A3F4A', alpha=0.9,
        fontproperties=mono_b, fontsize=36, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.123, y=1.113, s='population growth rate, per 1000', fontproperties=mono_r, color='#5D646F',
        fontsize=12, ha='center', va='bottom', transform=ax.transAxes)

# Added source of the data
ax.text(x=0.84, y=-0.068, s='Data: State Statistics Service of Ukraine', color='#5D646F', 
        fontproperties=mono_r, fontsize=12, alpha=1.0, ha='center', va='bottom', transform=ax.transAxes)

# Removed x and y axis labels
ax.set_xlabel('')
ax.set_ylabel('')

# Setting proper xticklabels
ax.set_xticklabels([1990, "", "", "", "", 1995, "", "", "", "", 2000, "", "", "", "", 2005, 
                    "", "", "", "", 2010, "", "", "", "", 2015]);

# Changing font family for x tick labels
for label in ax.get_xticklabels():
    label.set_fontproperties(mono_r)

# Changing font family for y tick labels
for label in ax.get_yticklabels():
    label.set_fontproperties(mono_r)
    
# Changing color of tick labels
plt.setp(ax.get_yticklabels(), color='#5D646F')
plt.setp(ax.get_xticklabels(), color='#5D646F');

# Setting proper font size for x tick labels
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(12)

# Setting proper font size for y tick labels
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(12)

# Put ticks inside (closer to axis)
ax.tick_params(direction='in')

# Added padding as new invisible axes
top_offset = -0.02
left_offset = 0.07
right_offset = -0.02
bottom_offset = 0.00
hgap = 0
ax_width = 1.02 - left_offset - right_offset
ax_height = (1.03 - top_offset - bottom_offset - hgap)
ax_add = fig.add_axes([left_offset, bottom_offset, ax_width, ax_height])

# Hide grid lines
ax_add.grid(False)

# Removing unnecessary tick marks
ax.tick_params(left=False, bottom=False, top=False, right=False)

# Hide axes ticks
ax_add.set_xticks([])
ax_add.set_yticks([])
ax_add.set_zorder(-1)

# Removing unnecessary box outline
ax_add.axis('off');
```
